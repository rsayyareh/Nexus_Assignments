# 2.pandas

```python
import pandas as pd
import numpy as np

# ساخت داده‌ تصادفی برای 30 روز
dates = pd.date_range(start="2025-01-01", periods=30)
temps = np.random.normal(20, 5, 30)      # دمای روزانه
rides = np.random.randint(50, 200, 30)   # تعداد رکاب‌ها
weekday = dates.day_name()               # استخراج نام روز هفته

# ساخت DataFrame
df = pd.DataFrame({
    "date": dates,
    "temp": temps,
    "rides": rides,
    "weekday": weekday
})
# ذخیره به فایل CSV
df.to_csv("rides.csv", index=False)
print("فایل rides.csv ساخته شد ✅")

df.head()
```
